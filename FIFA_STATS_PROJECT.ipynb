<span style="font-size:24px ; color:skyblue"> Displayed the top 20 players based on their overall rating using barplot.<span>



```python
top_20_players = df.sort_values(by='Overall',ascending=False).head(20)
plt.figure(figsize=(12,8))
sns.barplot(x='Overall',y='Name',data=top_20_players,palette='viridis')
plt.title('Top 20 Players Based on Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Player Name')
plt.show()
```

<span style="font-size:24 ; color:skyblue"> Displaying age distribution in various countries using violinplot.<span>

```python
selected_countries = ['England', 'Germany', 'Spain', 'Brazil', 'Argentina', 'Italy', 'France', 'Colombia', 'Japan', 'Netherlands']
df_selected_countries = df[df['Nationality'].isin(selected_countries)]

plt.figure(figsize=(14, 8))
sns.violinplot(x='Nationality', y='Age', data=df_selected_countries, palette='viridis')
plt.title('Age Distribution in Selected Countries (Violin Plot)')
plt.xlabel('Nationality')
plt.ylabel('Age')
plt.xticks(rotation=90)
plt.show()
```

```python
# List of most popular clubs
popular_clubs = ['Juventus', 'Liverpool', 'Ajax', 'Manchester City', 'Paris Sain-Germain',
                 'FC Bayern München', 'Real Madrid', 'FC Barcelona', 'Manchester United', 'FC Porto']

# Filter the DataFrame to include only the most popular clubs
df_popular_clubs = df[df['Club'].isin(popular_clubs)]

# Create a violin plot for the most popular clubs
plt.figure(figsize=(14, 8))
sns.violinplot(x='Club', y='Age', data=df_popular_clubs, palette='viridis')
plt.title('Age Distribution in Most Popular Clubs (Violin Plot)')
plt.xlabel('Club')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sort the DataFrame by age in descending order (oldest to youngest)
df_sorted_by_age = df.sort_values(by='Age', ascending=False)

# Print specific columns for the sorted DataFrame
columns_to_print = ['Name', 'Age', 'Nationality', 'Club']
print(df_sorted_by_age[columns_to_print])
```

```python
# Create a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age', y='Overall', data=df, alpha=0.5, color='skyblue')
plt.title('Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```

```python
# List of columns to create regression plots with
columns_to_plot = ['Stamina', 'Strength', 'Acceleration', 'SprintSpeed', 'Agility', 'Jumping', 'Vision', 'Reactions']

# Create regression plots for each specified column
plt.figure(figsize=(16, 12))
for column in columns_to_plot:
    plt.subplot(3, 3, columns_to_plot.index(column) + 1)
    sns.regplot(x='Age', y=column, data=df, scatter_kws={'alpha': 0.5, 'color': 'skyblue'})
    plt.title(f'Age vs {column}')

plt.tight_layout()
plt.show()
```
